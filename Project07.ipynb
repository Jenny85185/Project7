```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python

df = pd.read_csv('company_sales_data.csv')
df.head()
```

```python

df.set_index('month_number', inplace=True)
df_month_profit = df[['total_profit']]

print(df_month_profit)
```

```python
#Exercise 1: Read Total profit of all months and show it using a line plot
#Total profit data provided for each month. Generated line plot must include the following properties: –
#● X label name = Month Number
#● Y label name = Total profit

plt.plot(df_month_profit.index, df_month_profit['total_profit'])

plt.xlabel('Month Number')
plt.ylabel('Total Profit')

plt.title('Total Profit for Each Month')
plt.show()
```

```python
#Exercise 2: Get total profit of all months and show line plot with the following Style properties
#Generated line plot must include following Style properties: –
#● Line Style dotted and Line-color should be red
#● Show legend at the lower right location.
#● X label name = Month Number
#● Y label name = Sold units number
#● Add a circle marker.
#● Line marker color as read
#● Line width should be 3

plt.plot(df_month_profit.index, df_month_profit['total_profit'],
         linestyle=':',  
         color='red',    
         linewidth=3,   
         marker='o',     
         markerfacecolor='red',  
         markeredgewidth=2)  


plt.xlabel('Month Number')
plt.ylabel('Sold Units Number')
plt.title('Sold Units for Each Month')
plt.legend(['Sold Units'], loc='lower right')

plt.show()
```

```python
#Exercise 3: Read all product sales data and show it using a multiline plot
#Display the number of units sold per month for each product using multiline plots. (i.e., Separate Plotline
#for each product).
plt.plot(df.index, df['facecream'], label='Facecream', marker='o', linestyle='-',color="blue")
plt.plot(df.index, df['facewash'], label='Facewash', linestyle='--', linewidth=4, color="orange")
plt.plot(df.index, df['toothpaste'], label='Toothpaste', marker='s', linestyle='-.', color="green")
plt.plot(df.index, df['bathingsoap'], label='Bathingsoap', marker='^', linestyle=':',color="yellow")
plt.plot(df.index, df['shampoo'], label='Shampoo', marker='d', linestyle='-', color="violet")
plt.plot(df.index, df['moisturizer'], label='Moisturizer', marker='p', linestyle='-', color ="brown")


plt.xlabel('Month Number')
plt.ylabel('Units Sold')
plt.title('Units Sold Per Month for Each Product')
plt.legend(loc='lower right')



plt.show()
```

```python
#Exercise 4: Read toothpaste sales data of each month and show it using a scatter plot. Also, add a grid in the plot. gridline style should “–“
plt.scatter(df.index, df['toothpaste'], label='Toothpaste', marker='o', color="green")
plt.grid(True, linestyle='-')
plt.xlabel('Month Number')
plt.ylabel('Units Sold')
plt.title('Units Sold Per Month ')
plt.legend(loc='lower right')

plt.show()
```

```python
#Exercise 5: Read face cream and facewash product sales data and show it using the bar chart
plt.bar(df.index - 0.2, df['facecream'], width=0.4, label='Facecream', color='blue') 
plt.bar(df.index + 0.2, df['facewash'], width=0.4, label='Facewash', color='orange')  

plt.grid(True, linestyle='--')
plt.xlabel('Month Number')
plt.ylabel('Units Sold')
plt.title('Face Cream and Face Wash Sales per Month')
plt.legend()


plt.show()
```

```python
#Exercise 7: Read the total profit of each month and show it using the histogram to see the most common profit ranges
#count, bin_edges = np.histogram(df['total_profit'], 12)

df[['total_profit']].plot(kind='hist', bins=bin_edges, color='skyblue', edgecolor='black', rwidth=1)

plt.xlabel('Total Profit')
plt.ylabel('Frequency')
plt.title('Histogram of Total Profit for Each Month')


plt.xticks(bin_edges, rotation=45)


plt.tight_layout()  
plt.show()

```

```python
#Exercise 8: Calculate total sale data for last year for each product and show it using a Pie chart

total_sales = {
    'Facecream': (df['facecream'] * df['total_units']).sum(),
    'Facewash': (df['facewash'] * df['total_units']).sum(),
    'Toothpaste': (df['toothpaste'] * df['total_units']).sum(),
    'Bathingsoap': (df['bathingsoap'] * df['total_units']).sum(),
    'Shampoo': (df['shampoo'] * df['total_units']).sum(),
    'Moisturizer': (df['moisturizer'] * df['total_units']).sum(),
}

# Plotting the pie chart
labels = total_sales.keys()
sizes = total_sales.values()

plt.figure(figsize=(8, 8))
plt.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=140, colors=['blue','orange','green','yellow','violet','brown'])
plt.title('Total Sales Data for Each Product for Last Year')
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.show()
```

```python
#Exercise 9: Read Bathing soap facewash of all months and display it using the Subplot
fig, axes = plt.subplots(1, 2, figsize=(12, 6))

axes[0].plot(df.index, df['bathingsoap'], marker='o', color='yellow')
axes[0].set_title('Bathing Soap Sales')
axes[0].set_xlabel('Month Number')
axes[0].set_ylabel('Units Sold')
axes[0].grid(True, linestyle='--')


axes[1].plot(df.index, df['facewash'], marker='x', color='orange')
axes[1].set_title('Facewash Sales')
axes[1].set_xlabel('Month Number')
axes[1].set_ylabel('Units Sold')
axes[1].grid(True, linestyle='--')

plt.tight_layout()
plt.show()
```

```python
#Exercise 10: Read all product sales data and show it using the stack plot

products = ['facecream', 'facewash', 'toothpaste', 'bathingsoap', 'shampoo', 'moisturizer']
colors = ['blue', 'orange', 'green', 'yellow', 'violet', 'brown']  

plt.stackplot(df.index, df[products].T, labels=products, colors=colors, alpha=0.7)


plt.xlabel('Month Number')
plt.ylabel('Units Sold')
plt.title('Product Sales Per Month')
plt.legend(loc='upper left')


plt.show()
```

```python
#In this code, I'm recalculating the total_units for each month by summing the individual product sales (facecream, facewash, toothpaste, bathingsoap, shampoo, and moisturizer). 
#Then create a new DataFrame df_comparison that contains both the original total_units and the newly calculated calculated_total_units.
#After that, I compute the difference between these two values and add it as a new column called difference. 
#This was done because I noticed discrepancies between the visualized data and the provided total_units values on the graph.
df['calculated_total_units'] = df[['facecream', 'facewash', 'toothpaste', 'bathingsoap', 'shampoo', 'moisturizer']].sum(axis=1)

df_comparison = df[['total_units', 'calculated_total_units']].copy()

df_comparison.loc[:, 'difference'] = df_comparison['total_units'] - df_comparison['calculated_total_units']
df_comparison.index = range(1, len(df_comparison) + 1)
print(df_comparison)
```
